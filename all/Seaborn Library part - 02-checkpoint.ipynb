# Bar Plot in Seaborn Library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset("penguins")
```

```python
df
```

```python
df["island"]
```

```python
sns.barplot(x = df.island, y = df.bill_length_mm)
plt.show()
```

```python
sns.barplot (x = "island" , y = "bill_length_mm" , hue = "sex" ,  data = df)
plt.show()
```

# Order Parameter

```python
order1 = ["Dream"  ,"Torgersen",  "Biscoe"]
sns.barplot (x = "island" , y = "bill_length_mm" , hue = "sex" , data = df , order = order1)
plt.show()
```

# hea_order

```python
order1 = ["Dream"  ,"Torgersen",  "Biscoe"]
sns.barplot (x = "island" , y = "bill_length_mm" , hue = "sex" ,  data = df , order = order1 , hue_order = ["Female" , "Male"] )
plt.show()
```

```python
order1 = ["Dream" , "Torgersen" , "Biscoe"]
sns.barplot(x = "island" , y = "bill_length_mm" ,data = df,  hue="sex"  ,order = order1,  hue_order =["Female" , "Male"] )
plt.show()
```

```python
sns.barplot(x ="island" , y = "bill_length_mm" , data = df , hue = "sex" , palette = "gist_heat_r") 
plt.show()
```

# Saturation

```python
sns.barplot(x = "island" , y= "bill_length_mm" , data = df  , hue ="sex" , palette = "OrRd" , saturation= 0.2)
plt.show()
```

# Histogram In Seaborn Library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = sns.load_dataset("penguins")
```

```python
df1
```

```python
sns.displot(df1["flipper_length_mm"])
plt.show()
```

# bins 

```python
sns.displot(df1["flipper_length_mm"] , bins = [170,180,190,200,210,220,230,240])
plt.show()
```

# color

```python
sns.displot(df1["flipper_length_mm"], bins=[170,180,190,200,210,220,230,240] , color = "green")
plt.show()
```

```python
sns.displot(df1["flipper_length_mm"] , bins = [170,180,190,200,210,220,230,240] , kde = True , color = "red" )
plt.show()
```
